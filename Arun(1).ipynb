```python
# Mount Google Drive to access your data
from google.colab import drive
drive.mount('/content/drive')

# Install necessary libraries
!pip install segmentation-models-pytorch
!pip install albumentations

import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import segmentation_models_pytorch as smp
from tqdm import tqdm
import albumentations as A
from albumentations.pytorch import ToTensorV2
from sklearn.metrics import accuracy_score, jaccard_score
import time

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define paths
IMAGE_DIR = "/content/images"  # Adjust this path
MASK_DIR = "/content/masks"   # Adjust this path

class LandcoverDataset(Dataset):
    def __init__(self, image_dir, mask_dir, transform=None):
        self.image_dir = image_dir
        self.mask_dir = mask_dir
        self.transform = transform
        self.images = [f for f in os.listdir(image_dir) if f.endswith('.tif')]
        self.class_map = {
            0: 'Background',
            1: 'Building',
            2: 'Woodland',
            3: 'Water',
            4: 'Road'
        }

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_name = self.images[idx]
        img_path = os.path.join(self.image_dir, img_name)
        mask_path = os.path.join(self.mask_dir, img_name)

        image = np.array(Image.open(img_path))
        mask = np.array(Image.open(mask_path))

        if self.transform:
            augmentations = self.transform(image=image, mask=mask)
            image = augmentations["image"]
            mask = augmentations["mask"]

        return image, mask.long()

    def count_classes(self):
        class_counts = {cls: 0 for cls in self.class_map.values()}
        for mask_file in os.listdir(self.mask_dir):
            if mask_file.endswith('.tif'):
                mask_path = os.path.join(self.mask_dir, mask_file)
                mask = np.array(Image.open(mask_path))
                unique, counts = np.unique(mask, return_counts=True)
                for cls, count in zip(unique, counts):
                    class_counts[self.class_map[cls]] += count
        return class_counts

# Define transforms
train_transform = A.Compose([
    A.Resize(512, 512),  # Increased resolution
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.5),
    A.RandomRotate90(p=0.5),
    A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
    ToTensorV2(),
])

val_transform = A.Compose([
    A.Resize(512, 512),  # Increased resolution
    A.Normalize(mean=(0.485, 0.456, 0.406), std=(0.229, 0.224, 0.225)),
    ToTensorV2(),
])

# Create datasets
full_dataset = LandcoverDataset(IMAGE_DIR, MASK_DIR, transform=train_transform)

# Split dataset
train_size = int(0.8 * len(full_dataset))
val_size = len(full_dataset) - train_size
train_dataset, val_dataset = torch.utils.data.random_split(full_dataset, [train_size, val_size])

# Override transform for validation dataset
val_dataset.dataset.transform = val_transform

# Create dataloaders
train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=4, shuffle=False, num_workers=2)

def count_classes(mask_dir):
    unique_classes = set()
    for mask_file in os.listdir(mask_dir):
        if mask_file.endswith('.tif'):
            mask_path = os.path.join(mask_dir, mask_file)
            mask = np.array(Image.open(mask_path))
            unique_classes.update(np.unique(mask))
    return len(unique_classes)

# Count the number of classes
num_classes = count_classes(MASK_DIR)
print(f"Number of unique classes found: {num_classes}")
full_dataset = LandcoverDataset(IMAGE_DIR, MASK_DIR)
class_counts = full_dataset.count_classes()

# Visualize class distribution
plt.figure(figsize=(10, 6))
plt.bar(class_counts.keys(), class_counts.values())
plt.title('Distribution of Classes in the Dataset')
plt.xlabel('Classes')
plt.ylabel('Pixel Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('class_distribution.png')
plt.show()

for cls, count in class_counts.items():
    print(f'Number of {cls}: {count}')

# Define UNet++ model
model = smp.UnetPlusPlus(
    encoder_name="resnet34",
    encoder_weights="imagenet",
    in_channels=3,
    classes=num_classes,
)
model = model.to(device)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

# Training function
def train_epoch(model, loader, optimizer, criterion, device):
    model.train()
    total_loss = 0
    for images, masks in tqdm(loader):
        images = images.to(device)
        masks = masks.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, masks)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(loader)

# Validation function
def validate(model, loader, criterion, device):
    model.eval()
    total_loss = 0
    all_preds = []
    all_masks = []
    with torch.no_grad():
        for images, masks in tqdm(loader):
            images = images.to(device)
            masks = masks.to(device)

            outputs = model(images)
            loss = criterion(outputs, masks)

            total_loss += loss.item()

            preds = torch.argmax(outputs, dim=1)
            all_preds.extend(preds.cpu().numpy().flatten())
            all_masks.extend(masks.cpu().numpy().flatten())

    accuracy = accuracy_score(all_masks, all_preds)
    iou = jaccard_score(all_masks, all_preds, average='weighted')

    return total_loss / len(loader), accuracy, iou

# Training loop
num_epochs = 20
best_iou = 0
for epoch in range(num_epochs):
    start_time = time.time()
    train_loss = train_epoch(model, train_loader, optimizer, criterion, device)
    val_loss, val_accuracy, val_iou = validate(model, val_loader, criterion, device)
    end_time = time.time()

    epoch_time = end_time - start_time
    print(f"Epoch {epoch+1}/{num_epochs}, Time: {epoch_time:.2f}s, Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Val Accuracy: {val_accuracy:.4f}, Val IoU: {val_iou:.4f}")

    if val_iou > best_iou:
        best_iou = val_iou
        torch.save(model.state_dict(), 'best_model.pth')

# Load best model
model.load_state_dict(torch.load('best_model.pth'))

# Visualization and evaluation function
def visualize_and_evaluate(model, loader, device, num_samples=3):
    model.eval()
    fig, axes = plt.subplots(num_samples, 3, figsize=(15, 5*num_samples))

    metrics = {'iou': [], 'pixel_accuracy': []}

    with torch.no_grad():
        for i, (images, masks) in enumerate(loader):
            if i >= num_samples:
                break

            images = images.to(device)
            masks = masks.to(device)

            start_time = time.time()
            outputs = model(images)
            inference_time = time.time() - start_time

            preds = torch.argmax(outputs, dim=1)

            # Calculate metrics
            iou = jaccard_score(masks.cpu().numpy().flatten(), preds.cpu().numpy().flatten(), average='weighted')
            pixel_accuracy = accuracy_score(masks.cpu().numpy().flatten(), preds.cpu().numpy().flatten())

            metrics['iou'].append(iou)
            metrics['pixel_accuracy'].append(pixel_accuracy)

            # Visualize

            # Normalize the image to [0, 1] if it's a float, or ensure it's in [0, 255] if it's integer
            image = images[0].cpu().permute(1, 2, 0)
            if image.dtype == torch.float32 or image.dtype == torch.float64:
                 image = torch.clamp(image, 0, 1)  # Clamp float values to [0, 1]
            elif image.dtype == torch.uint8:
                 image = torch.clamp(image, 0, 255)  # Clamp integer values to [0, 255]

            # Now plot the normalized image
            axes[i, 0].imshow(image)
            axes[i, 0].set_title("Input Image")

            # Plot the masks and predictions
            axes[i, 1].imshow(masks[0].cpu(), cmap='nipy_spectral', vmin=0, vmax=num_classes-1)
            axes[i, 1].set_title("Ground Truth")
            axes[i, 2].imshow(preds[0].cpu(), cmap='nipy_spectral', vmin=0, vmax=num_classes-1)
            axes[i, 2].set_title("Prediction")

            print(f"Sample {i+1} - IoU: {iou:.4f}, Pixel Accuracy: {pixel_accuracy:.4f}, Inference Time: {inference_time:.4f}s")

    plt.tight_layout()
    plt.savefig('high_resolution_maps.png', dpi=300, bbox_inches='tight')
    plt.show()

    print(f"Average IoU: {np.mean(metrics['iou']):.4f}")
    print(f"Average Pixel Accuracy: {np.mean(metrics['pixel_accuracy']):.4f}")

# Visualize and evaluate results
visualize_and_evaluate(model, val_loader, device)

# Generate high-resolution maps
def generate_high_res_maps(model, loader, device, num_samples=3):
    model.eval()
    fig, axes = plt.subplots(num_samples, 1, figsize=(15, 10*num_samples))

    with torch.no_grad():
        for i, (images, _) in enumerate(loader):
            if i >= num_samples:
                break

            images = images.to(device)

            outputs = model(images)
            preds = torch.argmax(outputs, dim=1)

            # Visualize high-resolution map
            axes[i].imshow(preds[0].cpu(), cmap='nipy_spectral', vmin=0, vmax=num_classes-1)
            axes[i].set_title(f"High-Resolution Landcover Map {i+1}")
            axes[i].axis('off')

    plt.tight_layout()
    plt.savefig('high_resolution_landcover_maps.png', dpi=300, bbox_inches='tight')
    plt.show()

# Generate high-resolution maps
generate_high_res_maps(model, val_loader, device)

print("Performance Metrics and High-Resolution Maps have been generated and saved.")
```
